# **String Slicing**

```python
string = "Python Programming"

string[0:6]   # string[start_index : end_index : step_size]
string[:6]    # default start index is "0"
string[7:]    # default end index is "end of string" >> end index is not included.
string[2:5:]    # default step size is "1"
```

# **String Methods ( Mostly Use )**

# 1. split()

This method is used for convert a string into list.

```python
string = "My name is Hamza "

list1 = string.split(" ") # string.split(sep_str)

list2 = string.split() # by default it is " "

print(list1)
print(list2)
```

# 2. join()

This method is used for convert a list into string.

If any integer found in the list then error will occurs.

```python
list1 = ["My","name","is","Hamza"]

string1 = " ".join(list1)

string2 = " # ".join(list1)

print(string1)
print(string2)
```

```python
list1 = ["fg",4,"dfg",9,"df"]

string = " ".join(list1)

string
```

```python
string = "Python Programming"

len(string)
```

# 2. capitalize()

It will capitalize only first char of string or sentence.

```python
string = "Python Programming"

new_str = string.capitalize()

new_str
```

# 3. upper()

All char going to convert in capital case.

```python
string = "Python Programming"

new_str= string.upper()

new_str
```

# 4. lower()

It will convert string into lower case.

```python
string = "Python Programming"

new_str= string.lower()

new_str
```

# 5. casefold()

It will convert string into lower case as well as the non-english will also convert into lower case

```python
string = "Python ÃŸ Programming"

new_str= string.casefold()

new_str
```

# 6. swapcase()

It will convert all lowercase into uppercase and uppercase into lowercase in the string.

```python
string = "Python Programming"

new_str= string.swapcase()

new_str
```

# 7. title()

It will capitalize each first character of word in the string.

```python
string = "my name is hamza patel"

new_str= string.title()

new_str
```

# 8. strip()

It remove all whitespace at the beginning and at the end of the string.

```python
string = "       Python Programming     "

new_str= string.strip()

new_str
```

# 9. lstrip()

It remove all whitespace at the beginning of the string.

```python
string = "      Python Programming     "

new_str= string.lstrip()

new_str
```

# 10. rstrip()

It remove all whitespace at the end of string.

```python
string = "     Python Programming     "

new_str= string.rstrip()

new_str
```

# 11. replace()

It will convert old substring into new substring and store it in a new string.

```python
string = "Python Programming Python Programming"

new_str= string.replace("Python","C",1)   # string.replace( old_str , new_str , count )
print(string.replace("Python","C"))    # The count is optional. If count is not defined then it will replace all old_str to new_str.
print(new_str)
```

# 12. count()

It return how many times the value is appears in the string.

```python
string = "abcdacfndbcabd"

count_a= string.count("a")

count_a
```

# 13. index()

It returns the index position of substring.

```python
string = "abcdacfndbcabd"   # string.index( substring , start_index , end_index )

x= string.index("a")        # The start & end index is optional. If it not defined then strat_index = 0 and end_index is end of string.

y= string.index("a",3,7)

print(x,y)
```

# 14. find()

It returns the index position of substring and if substring is not found then it returns " **-1** "

```python
string = "abcdacfndbcabd"   # string.find( substring , start_index , end_index )

x= string.find("a")         # The start & end index is optional. If it not defined then strat_index = 0 and end_index is end of string.

y= string.find("a",3,7)

z= string.find("p")         # If the given substring is not found it returns "-1"

print(x,y,z)
```

# 15. isalnum()

This method check that string should contaning number's and char **[ a-z, A-Z, 0-9 ]**

It returns boolean **True** and **False**

If string has whitespace then also return value is "**False**"

```python
string = "datascience123"

ans = string.isalnum()

print(ans)

"data science".isalnum()
```

# 16. isalpha()

This method check that string should contaning char only **[ a-z, A-Z ]**

It returns boolean **True** and **False**

If string has whitespace then also return value is "**False**"

```python
string = "datascience"

ans = string.isalpha()

print(ans)

"data science".isalnum()
```

# 19. istitle()

If the first char of each word is capital then it returns "**True**"

```python
string = "Hamza Patel"

ans = string.istitle()

ans
```

# 20. isspace()

If the string contain only whitespace then it returns "**True**"

```python
string = "  "

ans = string.isspace()

ans
```

# 21. isdigit()

If string contains digits,subscripts,superscripts etc then it returns "**True**"

```python
string = "\u2082"

ans = string.isdigit()

ans
```

# 22. isdecimal()

If string contain only digits [ 0-9 ] then it returns "**True**"

Special char & whitespace are not allowed.

```python
ans1 = "123456".isdecimal()

ans2 = "123 456".isdecimal()

print(ans1,ans2)
```

# 23. isnumeric()

If all char are digits or subscripts or superscripts or fraction ( 1/2 , 3/4)  it returns "**True**"

Space & special char are not allowed.

```python
string = "\u00BD"

ans = string.isnumeric()

ans
```
