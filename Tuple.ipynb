# Q. What is Tuple ?

Tuple is used for store the multiple items in a single variable.

Data types >> **int , float , string , list , tuple , set , frozenset , complex , dict .**

# Properties Of Tuple

1. Tuple is ordered data type.

2. Tuple is immutable >> unchangable >> can not modify

3. Tuple also allow duplicate data

4. Tuple items also seprate by **comma ( , )**

5. Tuple data is represented by **paranthesis ( )**

# Diffrence between **List** & **Tuple**

          **Tuple**                       **List**
>> Immutable                     >> Mutable
>> faster compair to list        >> Slower compair to tuple
>> represented by ()             >> Represented by []

# **Immutable**

```python
tuple1 = (6,5.6,5,22,"abc",77,9,54,11,34,9,75,44)

tuple1[6]=88
```

# **Faster**

```python
import sys

tuple1 = (2,5.6,9,22,"string",77,9,54,11,34,9,75,44)

list1 = [2,5.6,9,22,"string",77,9,54,11,34,9,75,44]

size_of_tuple = sys.getsizeof(tuple1)

size_of_list = sys.getsizeof(list1)

print(f"size of tuple in byte is : {size_of_tuple}")
print(f"size of list in byte is : {size_of_list}")
```

# **Tuple Methods**

# 1. count()

It will **count** the **occurence** of an **data**.

```python
tuple1 = (6,5.6,5,22,"abc",90,5,65,12,43,5,75,56,5,98)

count_5 = tuple1.count(5)

count_5
```

# 2. index()

It will return index at first occurence of given data.

```python
tuple1 = (6,5.6,5,22,"abc",90,5,65,12,43,5,75,56,5,98)

x = tuple1.index(5,4,9)      # tuple1.index( data , start_index , end_index )

y = tuple1.index(5)      # The start & end index is optional. If it not defined then strat_index = 0 and end_index is end of string.

print(x,y)
```
