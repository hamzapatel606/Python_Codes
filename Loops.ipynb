# What is Loops ?

When we have to perform a task repeatedly for two or more time then we are using **Loop**

# **For Loop**

```python
Synatx:

sequence --- > string, list, tuple,range,set,dict ---> numpy array, ---> dataframe

for var in sequence:
    statements
    expression
    .....
```

```python
for char in "Hamza":
  print(char)
```

# **For else loop**

else loop will **applicable** when **all the iterations are perform** in for loop.

```python
for i in range(5):
  print(i)
else:
  print("All iteratins are perform.")
```

```python
for i in range(5):
  print(i)
  if i==3:
    break
else:
  print("All iteratins are perform.")
```

# **While Loop**

```python
Syntax:

while condition:
    statements
    expression
    ...

    ...
    increment condition or condition
```

```python
x = 0
while x<=10:
  print(x,end=" ")
  x=x+1
```

# **While else loop**

else loop will **applicable** when **all the iterations are perform** in while loop.

```python
x = 0
while x<5:
  print(x,end=" ")
  x=x+1
else:
  print("\nAll iterations are perform.")
```

```python
x = 0
while x<5:
  print(x,end=" ")
  if x==3:
    break
  x=x+1
else:
  print("\nAll iterations are perform.")
```

```python
1. break
2. continue
3. pass
```

# break

break statement is used to exit the loop.

```python
# stop the loop at index no . 5
for item in range(1,11):
    print(item,end=" ")
    if item == 5:
        print("\nThe loop has been terminated")
        break
```

# continue

Used to stop only current iteration and continue with next iteration.

```python
# stop the loop at index no . 5
for item in range(1,11):
    print(item,end=" ")
    if item == 5:
        continue

```
