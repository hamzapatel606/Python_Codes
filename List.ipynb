# What is list ?

List is the collection of item/data.

Data types >> **int , float , string , list , tuple , set , frozenset , complex , dict .**

# Properties Of List

1. Ordered data type

2. Mutable >> changeable >> **add , delete , update**

3. Duplicate data/item are allowed.

4. List items are separed by **comma ( , )**

List data is represented by **square bracket ( [] )**

```python
list1 = [ 5,6.8,"string",[1,5.8,"str"],(1,5.7,"abc"),{1,5,9,4},2+3j,{1:"a",2:"b",3:"c"}]

list1
```

# **List Methods**

# 1. append()

It will **add** an **item**/data at **end of list**.

```python
list1 = [1,3.9,8,55,29]

list1.append(77)

list1
```

# 2. extend()

This method is use for **add** an **iterable** at the **end of list**

```python
list1 = [1,4.6,"string",66]

list1.extend([1,4,8])

list1
```

# 3. count()

It will **count** the **occurence** of an **data**.

```python
list1 = [1,3,5,2,6,8,4,2,5,8,5,9,5,4,3,5]

count_5 = list1.count(5)

count_5
```

# 4. insert()

It will **insert value** at the given **index** in list.

```python
list1 = [1,3.5,"string",99,65,44,[1,6,9]]

list1.insert(4,"Hamza")     # list1.insert(index,value)

list1
```

# 5. index()

It will **return index** at **first occrunce** of given **data**.

```python
list1 = [2,5.6,9,22,"string",77,9,54,11,34,9,75,44]

x= list1.index(9,4,12)      # list1.index( data , start_index , end_index )

y= list1.index(9)      # The start & end index is optional. If it not defined then strat_index = 0 and end_index is end of string.

print(x,y)
```

# 6. remove()

It will **remove** the **first occurce** of the **data** from the list.

```python
list1 = [2,5.6,9,22,"string",77,9,54,11,34,9,75,44]

list1.remove(9)

list1
```

# 7. pop()

It will **remove** the data at the given **index**.

If the **index is not given** then it will **remove** the **last element** of the list.

```python
list1 = [2,5.6,9,22,"string",77,9,54,11,34,9,75,44]

list1.pop(4)

print(list1)

list1.pop()

print(list1)
```

# 8. clear()

It will **clear all the elements** of the **list**.

```python
list1 = [2,5.6,9,22,"string",77,9,54,11,34,9,75,44]

list1.clear()

list1
```

# 9. sort()

This method is used for **sort** a list in **ascending** or **decending order.**

```python
list1 = [2,5.6,9,22,77,9,54,11,34,9,75,44]
list1.sort()

list2 = [2,5.6,9,22,77,9,54,11,34,9,75,44]
list2.sort(reverse=True)

print(list1)
print(list2)
```

# **List Comprehension**

This is a **method** which is **used for making short the loops expression.**

```python
list1 = [ i for i in range(1,11)]   # list1 = [ expression for i in iterable ]

list2 = [ i**2 for i in range(1,11)]

print(list1)
print(list2)
```
