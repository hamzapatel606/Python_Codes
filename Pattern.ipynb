```python
n = 3
for i in range(1, n+1):
    print(" "*(n-i), end="")
    for j in range(i):
        print(chr(65+j), end="")
    for k in range(i):
        print(k+1,end="")
    print()

for i in range(1, n):
    print(" "*(i), end="")
    for j in range(n-i):
        print(chr(65+j), end="")
    for k in range(n-i):
        print(k+1,end="")
    print()
```
