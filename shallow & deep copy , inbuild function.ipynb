# **Shallow Copy**

In shallow copy we use the copy function.

If we made any change in sublist of list then it will reflect on anoter object's also.

```python
list1 = [[1,2,4],[3,6,8]]

list2 = list1.copy()

list1[1][1] = 55

print(list1)
print(list2)
```

# **Deep Copy**

In deep copy we use deepcopy function by importing copy labarary.

If we made any change in sublist of list then it does not reflect on another object's.

```python
import copy

list1 = [[1,2,5],[2,5,9]]

list2 = copy.deepcopy(list1)

list1[1][1] = 55

print(list1)
print(list2)
```

# **Equal sign ( = )**

If you copy the one object into another object by using equal to sign ( = ) then if you made any changes in any object it reflected in both object ( **Both objects has same memory location** )

```python
list1 = [1,3,5,7,2,9,5]

list2 = list1

list2[5] = 77

print(list1)
print(list2)
```

# **unpacked** ( * )

by using " * " we can unpacked any packed data type.

```python
list1 = [1,3,5,6,8,9,3,2,7,0,3]
tuple1 = (1,2,5,3,9,6,0,99)
set1 = {1,4,5,7,3,8,0,44,9}
dict1 = {1:"H",2:"a",3:"m",4:"z",5:"a"}
print(*list1)
print(*tuple1)
print(*set1)
print(*dict1)
```

# **del keyword**

By using **del** keyword we are deleting all the data of the object from database.

```python
a = 100

print(a)

del a

print(a)
```

# **Inbuild Functions**

# 1. len()

This method is used for find the length of any itrable.

```python
len1 = len([1,2,5,6,8,3,9,4,1,4,8])

len2 = len((1,3,5,1,6,7,2,9,4))

len3 = len({1,4,6,8,9,3,0,45,2,5,7})

len4 = len(range(6))

len5 = len("My name is Hamza Patel")

print(len1,len2,len3,len4,len5)
```

# 2. chr()

This method is returns the **character** of given **ASCII value**

```python
ans = chr(97)   # ans = chr( ascii_value)

print(ans)
```

# 3. ord()

This method is returns the **ASCII value** of given **char**.

```python
ans = ord("a")    #  ans = ord(char)

ans
```

# 4. enumerate()

This method is used when we require **index & value at index** in an iterable sequence.

```python
string = "Hamza Patel"

for string_index,char in enumerate(string):
  print(string_index,char)
```

# 5. sorted()

This method is **return** a **new sorted list**.

But the **old list** is **unchanged**.

```python
list1 = [1,4,6,9,23,4,56,88,10,21,77,43,-4,-66,-54]

list2 = sorted(list1)

print(list1)
print(list2)
```
